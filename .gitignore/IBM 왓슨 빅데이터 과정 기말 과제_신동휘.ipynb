```python
df_subway = pd.DataFrame(columns = ['역명', '승하차승객수'])

data = df1[(df1['노선명']=='선유도')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'선유도','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

data = df1[(df1['노선명']=='신목동')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'신목동','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

data = df1[(df1['노선명']=='염창')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'염창','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

data = df1[(df1['노선명']=='등촌')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'등촌','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

data = df1[(df1['노선명']=='증미')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'증미','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

data = df1[(df1['노선명']=='가양')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'가양','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

data = df1[(df1['노선명']=='양천향교')][['역명','승차총승객수']].sum()
data_sum = data.sum()
insert_data = {'역명':'양천향교','승하차승객수':data_sum}
df_subway = df_subway.append(insert_data, ignore_index=True)

df_subway
```

```python
import re
import requests
df_subway['역명'] = df_subway['역명'].apply(lambda x: re.sub(r'\([^)]*)','',x) if '(' in x else x)
df_subway['역명'] = df_subway['역명'].apply(lambda x: x+'역')

x=[]
y=[]
for query in df_subway['역명']:
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(query)
    apikey = '0407a2349ecda6d17bcefa82be4efd99'
    kakao= requests.get( url, params = {'query':query}, headers={'Authorization' : 'KakaoAK ' + apikey } ).json() 
    x.append(kakao['documents'][0]['y'])
    y.append(kakao['documents'][0]['x'])

df_subway['latitude'] =x
df_subway['longitude'] =y
df_subway
```

```python
map=folium.Map(location=[37.5507280806214, 126.865572139231], zoom_start=12)
for i in df_subway.index:
    folium.CircleMarker([df_subway['latitude'][i],df_subway['longitude'][i]],radius = df_subway['승하차승객수'][i]/2000,
                        popup=df_subway['역명'][i],color='#3186cc', fill_color='#3186cc').add_to(map)
map
```
